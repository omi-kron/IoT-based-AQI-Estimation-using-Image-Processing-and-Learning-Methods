```python

import numpy as np
import pandas as pd

def remove_timezone(dt):
   
    # HERE `dt` is a python datetime
    # object that used .replace() method
    return dt.replace(tzinfo=None)

def renaming_columns_avg(df1, avg = 'H'):
    
    df1['created_at'] = df1['created_at'].apply(remove_timezone)
    df1.set_index("created_at", inplace = True)
    df1 = df1.resample(avg, label='right', closed='right').mean().round(3)
    df1 = df1.dropna(axis=1, how='all')
    df1 = df1.drop(['entry_id'], axis = 1)
    df1 = df1.rename(columns = {'field1': 'PM10','field2':'PM2.5','field3':'RH','field4':'Temperature'})
    
    
    return df1
## PM2.5 Sub-Index calculation 
## PM2.5 is measured in ug/m3

def get_PM25_subindex(x):
    if x <= 30:
        return x * 50 / 30
    elif x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0


## PM10 Sub-Index calculation
## PM10 is measured in ug/m3

def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x <= 100:
        return x
    elif x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x <= 350:
        return 200 + (x - 250)
    elif x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0


## AQI bucketing
def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN
    
df = pd.read_csv('Statics_nodes//Monthwise_data//Dec_pm10maxcv_all.csv')
#df = renaming_columns_avg(df)
#df = df.dropna(subset ={'PM10'})

df["PM2.5_SubIndex"] = df["PM2.5"].apply(lambda x: get_PM25_subindex(x))

df["PM10_SubIndex"] = df["PM10"].apply(lambda x: get_PM10_subindex(x))




df["Checks"] = (df["PM2.5_SubIndex"] > 0).astype(int) + \
                (df["PM10_SubIndex"] > 0).astype(int) 

df["AQI_calculated"] = round(df[["PM2.5_SubIndex", "PM10_SubIndex"]].max(axis = 1))
#df.loc[df["PM2.5_SubIndex"] + df["PM10_SubIndex"] <= 0, "AQI_calculated"] = np.NaN
#df.loc[df.Checks < 3, "AQI_calculated"] = np.NaN

df["AQI_bucket_calculated"] = df["AQI_calculated"].apply(lambda x: get_AQI_bucket(x))
df[~df.AQI_calculated.isna()].head(13)

df
```
